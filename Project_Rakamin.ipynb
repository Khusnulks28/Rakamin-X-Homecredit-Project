```python
import numpy as np

# Menghitung IQR untuk AMT_INCOME_TOTAL
Q1_AMT_INCOME_TOTAL = data_new['AMT_INCOME_TOTAL'].quantile(0.25)
Q3_AMT_INCOME_TOTAL = data_new['AMT_INCOME_TOTAL'].quantile(0.75)
IQR_AMT_INCOME_TOTAL = Q3_AMT_INCOME_TOTAL - Q1_AMT_INCOME_TOTAL

# Batas bawah dan atas
lower_bound_AMT_INCOME_TOTAL = Q1_AMT_INCOME_TOTAL - 1.5 * IQR_AMT_INCOME_TOTAL
upper_bound_AMT_INCOME_TOTAL = Q3_AMT_INCOME_TOTAL + 1.5 * IQR_AMT_INCOME_TOTAL

# Menghapus outlier (opsional: filter data dalam rentang batas)
data_new["Clipped_AMT_INCOME_TOTAL"] = data_new["AMT_INCOME_TOTAL"].clip(lower_bound_AMT_INCOME_TOTAL, upper_bound_AMT_INCOME_TOTAL)

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Memisahkan fitur (X) dan target (y)
X = hasil_df.drop(columns=['TARGET'])  # Semua fitur kecuali target
y = hasil_df['TARGET']  # Target untuk prediksi

# Membagi dataset menjadi training & testing set (80:20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2021, stratify=y)
```

### Modelling

#### Logistic Regression

```python
# Inisialisasi model Logistic Regression
logreg = LogisticRegression(max_iter=1000, random_state=42)

# Melatih model
logreg.fit(X_train, y_train)

# Melakukan prediksi
y_pred = logreg.predict(X_test)
y_pred_prob = logreg.predict_proba(X_test)[:, 1]  # Probabilitas untuk kelas positif
```

```python
# Akurasi Model
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi Model: {accuracy:.4f}")

# AUC-ROC Score
roc_auc = roc_auc_score(y_test, y_pred_prob)
print(f"AUC-ROC Score: {roc_auc:.4f}")

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['No Default', 'Default'], yticklabels=['No Default', 'Default'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Inisialisasi Model Random Forest
rf = RandomForestClassifier(n_estimators=200, max_depth=20, random_state=100, class_weight='balanced')

# Melatih Model
rf.fit(X_train, y_train)

# Prediksi
y_pred_rf = rf.predict(X_test)
y_pred_prob_rf = rf.predict_proba(X_test)[:, 1]

# Akurasi Model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Akurasi Model: {accuracy_rf:.4f}")

# AUC-ROC Score
roc_auc_rf = roc_auc_score(y_test, y_pred_prob_rf)
print(f"AUC-ROC Score: {roc_auc_rf:.4f}")

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred_rf))

# Confusion Matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(5, 4))
sns.heatmap(cm_rf, annot=True, fmt="d", cmap="Blues", xticklabels=['No Default', 'Default'], yticklabels=['No Default', 'Default'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Random Forest")
plt.show()
```
