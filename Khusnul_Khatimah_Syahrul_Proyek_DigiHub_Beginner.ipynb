### b.) Clipping ( Handle Outlier )

```python
df.describe()
```

insight ( Praduga ) :    

1. Kolom website_recency kemungkinan terdapat high outlier karena rentangnya adalah dari 0 - 93 namun rata-ratanya adalah 20.

2. Kolom website_purchase_amount kemungkinan terdapat outlier karena rentangnya adalah dari 0 - 9.5jt namun rata-ratanya adalah 866rb dengan median 328rb.

Info tambahan :    

e adalah eksponen atau dikali dengan 10 pangkat. jadi e+05 artinya dikali 10 pangkat 5. yang berarti, rentang website_purchase_amount adalah dari 0 - 9.500.000 dan mean nya adalah 866.000

Handle outlier website_recency :

```python
import numpy as np

# Menghitung IQR untuk website_recency
Q1_recency = df['website_recency'].quantile(0.25)
Q3_recency = df['website_recency'].quantile(0.75)
IQR_recency = Q3_recency - Q1_recency

# Batas bawah dan atas
lower_bound_recency = Q1_recency - 1.5 * IQR_recency
upper_bound_recency = Q3_recency + 1.5 * IQR_recency

# Menghapus outlier (opsional: filter data dalam rentang batas)
df["Clipped_website_recency"] = df["website_recency"].clip(lower_bound_recency, upper_bound_recency)

# Verifikasi
if df['website_recency'].min() >= lower_bound_recency and df['website_recency'].max() <= upper_bound_recency:
    print("Kolom website_recency sudah tidak ada outlier")
else:
    print("Kolom website_recency masih memiliki outlier")
```

Kolom website_recency sudah tidak ada outlier

Handle outlier website_purchase_amount

```python
# Menghitung IQR untuk website_purchase_amount
Q1_purchase = df['website_purchase_amount'].quantile(0.25)
Q3_purchase = df['website_purchase_amount'].quantile(0.75)
IQR_purchase = Q3_purchase - Q1_purchase

# Batas bawah dan atas
lower_bound_purchase = Q1_purchase - 1.5 * IQR_purchase
upper_bound_purchase = Q3_purchase + 1.5 * IQR_purchase

# Meng-clipping nilai outlier
df["Clipped_website_purchase_amount"] = df["website_purchase_amount"].clip(lower_bound_purchase, upper_bound_purchase)


# Statistik deskriptif kolom yang sudah di-clipping
print(df["Clipped_website_purchase_amount"].describe())
```

kolom website_purchase_amount yang baru adalah rentang dari 0 - 1.4jt dengan rata-ratanya 437rb ( make sense )

Encoding kolom province_group dan age_group

```python
#PROVINCE
# Membuat dictionary berdasarkan kelompok provinsi
kelompok_provinsi = {
    "indo_barat": [
        "Aceh", "Sumatera Utara", "Sumatera Barat", "Riau", "Kepulauan Riau",
        "Jambi", "Sumatera Selatan", "Bangka Belitung", "Bengkulu", "Lampung",
        "Jakarta", "Jawa Barat", "Banten", "Jawa Tengah",
        "DI Yogyakarta", "Jawa Timur", "Kalimantan Barat"
    ],
    "indo_tengah": [
        "Kalimantan Tengah", "Kalimantan Selatan", "Kalimantan Timur",
        "Kalimantan Utara", "Bali", "Nusa Tenggara Barat (NTB)",
        "Nusa Tenggara Timur (NTT)", "Sulawesi Utara", "Sulawesi Tengah",
        "Sulawesi Selatan", "Sulawesi Tenggara", "Gorontalo", "Sulawesi Barat"
    ],
    "indo_timur": [
        "Maluku", "Maluku Utara", "Papua", "Papua Barat", "Papua Tengah",
        "Papua Pegunungan", "Papua Selatan", "Kepulauan Aru",
        "Kepulauan Raja Ampat"
    ]
}


# Membuat fungsi untuk menentukan kelompok provinsi
def assign_province_group(province):
    for group, provinces in kelompok_provinsi.items():
        if province in provinces:
            return group
    return "unknown"  # Jika tidak cocok dengan kelompok manapun

# Menambahkan kolom baru berdasarkan fungsi assign_province_group
df['province_group'] = df['province'].apply(assign_province_group)


# Menampilkan hasil
print(df[['province', 'province_group']])



```
